```python
import pandas as pd
```

```python
data = pd.read_csv(
    "./intents/raw_data/20000-Utterances-Training-dataset-for-chatbots-virtual-assistant-Bitext-sample.csv")

data[
    (data["intent"] == "get_invoice")
    | (data["intent"] == "get_refund")
    | (data["intent"] == "payment_issue")
]["utterance"].to_csv('./intents/payments.csv', index=False, header=[0])

# delivery_options delivery_period set_up_shipping_address
data[
    (data["intent"] == "delivery_options")
    | (data["intent"] == "delivery_period")
    | (data["intent"] == "set_up_shipping_address")
]["utterance"].to_csv('./intents/delivery.csv', index=False, header=[0])
```
